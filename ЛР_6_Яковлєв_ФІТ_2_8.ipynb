```python
import numpy as np
import matplotlib.pyplot as plt

x_points = np.array([-3, -2, -1, 0])
f_points = np.array([4, 4, 0, -2])

def lagrange_interpolation(x, x_points, f_points):
    n = len(x_points)
    L = 0
    for i in range(n):
        term = f_points[i]
        for j in range(n):
            if i != j:
                term *= (x - x_points[j]) / (x_points[i] - x_points[j])
        L += term
    return L

x_values = np.array([-4, -3.5, 0.5, 1])

f_values = lagrange_interpolation(x_values, x_points, f_points)

for x, f in zip(x_values, f_values):
    print(f"f({x}) ≈ {f:.4f}")

x_plot = np.linspace(-4, 1, 500)
y_plot = lagrange_interpolation(x_plot, x_points, f_points)

plt.plot(x_points, f_points, 'o', label='Задані точки')
plt.plot(x_plot, y_plot, label="Інтерполяційна функція")
plt.scatter(x_values, f_values, color='red', label="Обчислені значення")
plt.legend()
plt.xlabel('x')
plt.ylabel('f(x)')
plt.title('Інтерполяційний многочлен Лагранжа')
plt.grid(True)
plt.show()
```
